```python
# integers

num1 = int(input("Enter the first number here :- ")) # 1 = "1" = 1
num2 = int(input("Enter the second number here :- "))# 2 = "2" = 2

sum = num1 + num2 # "2"+"1" = 21 ; 1+2 = 3
print(sum)
```

```python
# Float are decimal numbers

num1 = float(input("Enter the first number here :- "))
num2 = float(input("Enter the second number here :- "))

sum = num1+num2
print(sum)
```
